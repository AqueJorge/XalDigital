```python
import json
import requests
import pandas as pd
```

```python
pd.set_option('max_columns', None)
```

```python
URL = "https://api.stackexchange.com/2.2/search?order=desc&sort=activity&intitle=perl&site=stackoverflow"
```

## 1.- Connecting with to url

```python
#Obtaining the data from url

def Data_From_URL(URL):
    response = requests.get(URL)
    result = response.json()
    return result
```

```python
#Calling the function to pass the URL
url_data = Data_From_URL(URL)
```

```python
#Getting the data
url_data
```

```python
#Using a DF to manipulate the data easier
url_df = pd.json_normalize(url_data['items'])
#Displaying the data in DF form
url_df
```

```python
# Information about the data
url_df.info()
```

```python
#Number of rows and columns
print("Rows #:", url_df.shape[0])
print("Columns #:",url_df.shape[1])
```

```python
#Sum of null values by column
url_df.isna().sum()
```

## 4.- The oldest and newest answer

```python
#Changing the format of the creation date
url_df.creation_date = pd.to_datetime(url_df.creation_date, unit = 's')

#Obtaining the oldest and newest Answers

oldest = url_df[url_df["link"]== url_df.creation_date.min()]

display("THE OLDEST ANSWER WAS AT:", url_df.creation_date.min(), url_df[url_df.creation_date == url_df.creation_date.min()])
display("----------------------------------------------------------------------")
display("THE NEWEST ANSWER WAS AT:", url_df.creation_date.max(), url_df[url_df.creation_date == url_df.creation_date.max()])
```
